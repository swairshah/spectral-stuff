```python
from gram_schmidt import *
from matrix_stuff import *
from conjugate_gs import *
from laplacian import *
%matplotlib inline

tol = 1e-6
plt.figure(figsize=(10,10))
G = GenDumbbellGraph(4,3)
pos = nx.fruchterman_reingold_layout(G)
L = Laplacian(G)
E = IncidenceMatrix(G)
Draw(G)
```

```python
scores = np.diag(E.T @ L @ E)
highest = -np.sort(np.unique(-scores))
print(highest)
idx = np.argsort(-scores)
print(scores[idx[0]])
print(np.count_nonzero(scores == highest[0]))
E_s = E[:,(scores == highest[0])]
G_s = GraphFromIncidenceMatrix(E_s)
plt.figure(figsize=(10,10))
Draw(G_s, pos)
```

```python
scores = np.diag(E.T @ L @ E)
highest = -np.sort(np.unique(-scores))
idx = np.argsort(-scores)
e = E[:,idx[0]]
print(idx[0])
for f in E.T:
    print(f, end=' ')
    denom = f.T @ L @ f
    if denom > tol:
        numer = f.T @ L @ e
        f = f - (numer/denom)*e
        print(np.linalg.norm(f), f)
```

```python
k = 5
P, D = conjugate_gs(E, L)
print(D)
idx = np.argsort(-D)
print(idx)
E_h = E[:, idx[:k]]
H = GraphFromIncidenceMatrix(E_h)
Draw(H, pos)
```
